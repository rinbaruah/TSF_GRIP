# Introduction

In this particular task of our internship, we will be focussing on the linear regression problem using the data set provided to us.

The data set contains the number of study hours and the scores obtained by the student. According to our intuition, the more a student studies, the more the marks should be obtained by the student. However, this is only a hypothesis so far and we will have to speak with our data if we are suppose to make such a statement. So, we shall make a few visualisations to see if we are gonna prove our hypothesis correct.

# Step 1: Importing relevant libraries and the dataset

Alright ! The first and foremost step should be to import the required libraries and the dataset. Let us do that right away and start off our analysis.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
%matplotlib inline
```

```python
link='http://bit.ly/w-data'
df=pd.read_csv(link)
```

```python
df.head()
```

As we can see from the above dataframe, we have two columns contaning the number of hours and scores. Hence, we will be trying to predict the scores of the student based on just one single feature which is number of study hours. Such a type of linear regression is called **Simple Linear Regression** that only considers the use of one feature to predict the result.

# Step 2: Data Visualisation

In this step, we will try to understand how are our study hours and scores distributed exactly. This will help us understand what the mean and median values are. Let's do this !

Using the pandas **describe** function will give us a quick look at the various indicators of our feature.

```python
df.describe()
```

From the above dataframe, we can see various indicators of both Study hours and Scores. The dataframe is self explanatory. Hence, what we shall do is try to replicate the above numbers into visualisations.

```python
fig=plt.figure(figsize=(20,10))
ax1=fig.add_subplot(121)
ax2=fig.add_subplot(122)
sns.boxplot('Hours',data=df,ax=ax1,orient='v',color='red')
sns.boxplot('Scores',data=df,ax=ax2,orient='v',color='green')
ax1.set_ylabel('Hours',size=20)
tick_1=np.arange(11)
ax1.set_yticklabels(tick_1,fontsize=15)
ax1.set_title('Study hours \n \n Median:{}'.format(df['Hours'].median()),size=20)

ax2.set_ylabel('Scores',size=20)
ax2.set_title('Scores  \n \n Median:{}'.format(df['Scores'].median()),size=20)

tick_2=np.arange(10,110,10)
ax2.set_yticklabels(tick_2,fontsize=15)

```

The above visualisations are called boxplots which shows almost every term in the dataframe we just checked. Values within the red and green boxes show the inter-quartile range. These are values between 25 percentile and 75 percentile. The end horizontal lines are called whiskers. All values beyond these whiskers are called outliers.


The horizontal line within the the box shows the median value.

The median number of study hours is 4.8 hours while median scores are 47.

Let us use another visualisation to see how the study hours and scores are distributed.

```python
fig2=plt.figure(figsize=(20,10))
ax1=fig2.add_subplot(121)
ax2=fig2.add_subplot(122)
g=sns.distplot(df['Hours'],ax=ax1)
h=sns.distplot(df['Scores'],ax=ax2,color='seagreen')
ax1.set_title('Study hours distribution',size=15)
ax2.set_title('Scores distribution',size=15)

g.axvline(df['Hours'].median(),color='red',label='Median study hour')
h.axvline(df['Scores'].median(),color='black',label='Median study hour')
h.legend()
g.legend()
```

From the distplot above, we can see that a high density of scores are between 20-40. The scores at either extremeties are low.

In study hours too, high density of study hours range from 1-3 hours.


Now that we have visualised the distributions, let us see how the study hours are related to scores through a regplot.

```python
plt.figure(figsize=(10,8))
sns.regplot(df['Hours'],df['Scores'],line_kws={'color':'red','linewidth':2.5})
plt.xlabel('Hours',size=15)
plt.ylabel('Scores',size=15)
plt.title('Relation between scores and study hours',size=20)
```

As expected, we have a linear relation between scores and study hours. Hence, statistically, our hyptothesis has been satisfied. More study hours lead to more test scores.


# Step 3: Performing regression analysis for prediction

Now that we have seen a linear relationship, let us perform a linear regression to find the value of scores at 9.25 study hours.

```python
from sklearn.model_selection import train_test_split
X=df['Hours']
y=df['Scores']
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,shuffle=True,random_state=0)
```

```python
reg_lin=LinearRegression()
reg_lin.fit(X_train.values.reshape(-1,1),y_train.values.reshape(-1,1))
```

```python
reg_lin.score(X_train.values.reshape(-1,1),y_train.values.reshape(-1,1))
```

```python
y_pred=reg_lin.predict(X_test.values.reshape(-1,1))
```

```python
reg_lin.score(X_test.values.reshape(-1,1),y_test.values.reshape(-1,1))
```

```python
inp_hrs=np.array(9.25).reshape(1,-1)
```

```python
reg_lin.predict(inp_hrs)
```

```python
df_comp=pd.DataFrame(y_test)
df_comp.reset_index(drop=True,inplace=True)
df_comp['Predictions']=y_pred
df_comp['Absolute error']=abs(df_comp['Scores']-df_comp['Predictions'])
df_comp['% error']=100*(df_comp['Absolute error'])/df_comp['Scores']
```

```python
df_comp
```

Let us try to visualise our results on a regplot. For an absolutely correct prediction, we should have all our points aligned along a linear line at 45 degrees. Let us see how our predictions differ from actual values.

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error

plt.figure(figsize=(10,8))
sns.regplot('Predictions','Scores',data=df_comp,
            line_kws={'color':'green','linewidth':2.5},
            scatter_kws={'color':'red','s':50},marker='*')

rmse_preds=np.sqrt(mean_squared_error(y_pred,y_test))

mae=mean_absolute_error(y_pred,y_test)
plt.xlabel('Predictions',size=15)
plt.ylabel('Actual scores',size=15)

plt.title('Prediction vs Actual scores  \n \n RMSE: {0:.2f} \n Mean Absolute Error: {1:.2f}'.format(rmse_preds,mae),size=25)
```

```python
print('Predicted score using Linear Regression with 9.25 hours of study: {0:.2f}'.format(reg_lin.predict(inp_hrs)[0][0]))
```

# Step 4: Conclusion

Once we have applied a simple linear regression, we could draw the following conclusions:

* The root mean squared error of the predictions is 4.65
* The mean absolute error is 4.18
* The predicted score of a student who studies for 9.25 hours is **93.69** according to the simple linear regression algorithm
